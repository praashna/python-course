```python
# function with argument and no return type
def cal(l,b): #parameter
    a = l*b
    print(a)
    
l = int(input("enter l = "))
b = int(input("enter b = "))
    
cal(l,b) #argument

```

```python
# function with argument and no return type

def area(x,y): #parameter
    a = x*y
    print(a)

def volume(x,y,z):
    v = x*y*z
    print(v)
    
l = int(input("enter l = "))
b = int(input("enter b = "))
h = int(input("enter h = "))
    
area(l,b) #argument
volume(l,b,h)
```
